```python
# Q2
np.random.seed(123)
MAT = np.random.exponential(scale=1/5, size=(10, 7))
MAT= np.around(MAT, decimals=3 )
```

```python
# 2-1
Q1 = np.percentile(MAT, 25)
Q3 = np.percentile(MAT, 75)

n1 = Q1-1.5*(Q3-Q1)
n2 = Q3+1.5*(Q3-Q1) 

MAT[(MAT<n1) | (MAT>n2)]
```

```python
# 2-2
Transformed = np.abs(MAT - np.mean(MAT))
closest = np.min(Transformed)
closest
row, col = np.where(Transformed == closest)
print(row, col, MAT[[row, col]])
```
